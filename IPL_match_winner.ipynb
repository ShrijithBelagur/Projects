```python
raw_df.dropna(inplace = True)
```

```python
teams = list(enumerate(('Mumbai Indians', 'Sunrisers Hyderabad', 'Kolkata Knight Riders',
       'Rajasthan Royals', 'Chennai Super Kings', 'Delhi Capitals',
       'Punjab Kings', 'Kings XI Punjab', 'Delhi Daredevils',
       'Rising Pune Supergiant', 'Gujarat Lions',
       'Rising Pune Supergiants', 'Pune Warriors', 'Deccan Chargers',
       'Kochi Tuskers Kerala', 'Royal Challengers Bangalore')))

teams = dict((j,i) for i,j in teams)

stadiums = list(enumerate(('MA Chidambaram Stadium', 'Wankhede Stadium',
       'Narendra Modi Stadium', 'Arun Jaitley Stadium',
       'Zayed Cricket Stadium', 'Sheikh Zayed Stadium', 'Eden Gardens',
       'Sawai Mansingh Stadium', 'Feroz Shah Kotla',
       'M Chinnaswamy Stadium', 'Rajiv Gandhi International Stadium',
       'Punjab Cricket Association IS Bindra Stadium',
       'Andhra Cricket Association-Visakhapatnam District Cricket Association Stadium',
       'Maharashtra Cricket Association Stadium',
       'Holkar Cricket Stadium', 'Saurashtra Cricket Association Stadium',
       'Green Park',
       'Dr. Y.S. Rajasekhara Reddy ACA-VDCA Cricket Stadium',
       'Shaheed Veer Narayan Singh International Stadium',
       'Sardar Patel Stadium', 'Punjab Cricket Association Stadium',
       'Brabourne Stadium', 'JSCA International Stadium Complex',
       'Barabati Stadium', 'Subrata Roy Sahara Stadium',
       'Himachal Pradesh Cricket Association Stadium', 'Nehru Stadium',
       'Dr DY Patil Sports Academy',
       'Vidarbha Cricket Association Stadium', 'Newlands',
       "St George's Park", 'Kingsmead', 'SuperSport Park', 'Buffalo Park',
       'New Wanderers Stadium', 'De Beers Diamond Oval',
       'OUTsurance Oval')))

stadiums = dict((j,i) for i,j in stadiums)
```

```python
class databuilder:
  def __init__ (self,dfname,team):
    self.team = team

    self.dfname = raw_df.loc[((raw_df['team1'] == f'{team}') | (raw_df['team2'] == f'{team}'))]
    self.dfname.drop(columns =['margin','man_of_the_match','winner','place'], axis=1, inplace=True)
    self.dfname.reset_index(inplace = True)

    team_score = []
    opponent_team_score = []

    for i in range(len(self.dfname)):
      if self.dfname['team1'].values[i] == f'{team}':
        team_score.append(self.dfname['team1_score'].values[i])
        opponent_team_score.append(self.dfname['team2_score'].values[i])
      else:
        team_score.append(self.dfname['team2_score'].values[i])
        opponent_team_score.append(self.dfname['team1_score'].values[i])

    self.dfname.drop(columns =['team1_score','team2_score','index'], axis=1, inplace=True)

    for i in range(len(self.dfname)):
      if self.dfname['toss_winner'].values[i] == f'{team}':
        self.dfname['toss_winner'].values[i] = 1
      else:
        self.dfname['toss_winner'].values[i] = 0

    opponent_team = []

    for i in range(len(self.dfname)):
      if self.dfname['team1'].values[i] != f'{team}':
        opponent_team.append(self.dfname['team1'].values[i])
      else:
        opponent_team.append(self.dfname['team2'].values[i])

    self.dfname['Opponent_Team'] = opponent_team

    self.dfname.drop(columns =['team1','team2'], axis=1, inplace=True)

    self.dfname['score'] = team_score
    self.dfname['Opponent_Team_score'] = opponent_team_score

    self.dfname.dropna(inplace = True)

    for i in range(len(self.dfname)):
      if self.dfname['toss_choice'].values[i] == 'Bat':
        self.dfname['toss_choice'].values[i] = 1
      else:
        self.dfname['toss_choice'].values[i] = 0

    for i in range(len(self.dfname)):
      self.dfname['Opponent_Team'].values[i] = teams[self.dfname['Opponent_Team'].values[i]]
      self.dfname['stadium'].values[i] = stadiums[self.dfname['stadium'].values[i]]



  def print(self):
    return self.dfname
```

```python
CSK = databuilder('df1', 'Chennai Super Kings')
MI = databuilder('df2', 'Mumbai Indians')
KKR = databuilder('df3', 'Kolkata Knight Riders')
PK = databuilder('df4', 'Punjab Kings')
DC = databuilder('df5', 'Delhi Capitals')
RR = databuilder('df6', 'Rajasthan Royals')
SRH = databuilder('df7', 'Sunrisers Hyderabad')
KXIP = databuilder('df8', 'Kings XI Punjab')
DD = databuilder('df9', 'Delhi Daredevils')
GL = databuilder('df10', 'Gujarat Lions')
RPS = databuilder('df11', 'Rising Pune Supergiant')
PW = databuilder('df12', 'Pune Warriors')
KTK = databuilder('df14', 'Kochi Tuskers Kerala')
RCB = databuilder('df15', 'Royal Challengers Bangalore')
```

#Model building and training

```python
import numpy as np
```

```python
param = {
    ('Chennai Super Kings') :[]
    ,('Mumbai Indians') :[]
    ,('Kolkata Knight Riders') :[]
    ,('Punjab Kings') :[]
    ,('Delhi Capitals') :[]
    ,('Rajasthan Royals') :[]
    ,('Sunrisers Hyderabad') :[]
    ,('Kings XI Punjab') :[]
    ,('Delhi Daredevils') :[]
    ,( 'Gujarat Lions') :[]
    ,( 'Rising Pune Supergiant') :[]
    ,( 'Pune Warriors') :[]
    ,( 'Kochi Tuskers Kerala') :[]
    ,( 'Royal Challengers Bangalore') :[]
}

modelacc = {
    ('Chennai Super Kings') :[]
    ,('Mumbai Indians') :[]
    ,('Kolkata Knight Riders') :[]
    ,('Punjab Kings') :[]
    ,('Delhi Capitals') :[]
    ,('Rajasthan Royals') :[]
    ,('Sunrisers Hyderabad') :[]
    ,('Kings XI Punjab') :[]
    ,('Delhi Daredevils') :[]
    ,( 'Gujarat Lions') :[]
    ,( 'Rising Pune Supergiant') :[]
    ,( 'Pune Warriors') :[]
    ,( 'Kochi Tuskers Kerala') :[]
    ,( 'Royal Challengers Bangalore') :[]
}
```

```python
class iterator:
  def __init__(self, dfname, team):
    self.team = team
    self.dfname = dfname

  def train(self, epoch):

    w = np.random.rand(5,2)

    b = np.array([0.0,0.0], ndmin=2)

    learning_rate = 0.001

    for j in range(epoch):

       for i in range(int(0.9*len(self.dfname.print()))):

        input = np.array([teams[self.team], self.dfname.print().Opponent_Team[i], self.dfname.print().toss_winner[i],  self.dfname.print().toss_choice[i], self.dfname.print().stadium[i]], ndmin=2)
        y = np.array([self.dfname.print().score[i], self.dfname.print().Opponent_Team_score[i]], ndmin=2)

        output = np.matmul(input, w) + b

        mse = (y-output)**2

        dw = -2*(np.matmul(np.transpose(input),(y-output)))

        db = -2*(y-output)

        w -= learning_rate*dw

        b -= learning_rate*db

    param[self.team] = [w, b]

  def validate(self):

    acc = 0

    for i in range(int(0.8*len(self.dfname.print())), len(self.dfname.print())):

      input = np.array([teams[self.team], self.dfname.print().Opponent_Team[i], self.dfname.print().toss_winner[i],  self.dfname.print().toss_choice[i], self.dfname.print().stadium[i]], ndmin=2)
      y = np.array([self.dfname.print().score[i], self.dfname.print().Opponent_Team_score[i]], ndmin=2)

      w, b = param[self.team]

      output = np.matmul(input,w) + b

      mse = (y-output)**2

      if self.dfname.print().score[i] > self.dfname.print().Opponent_Team_score[i] and output[0][0] > output[0][1]:
        acc = acc + 1
      elif self.dfname.print().score[i] < self.dfname.print().Opponent_Team_score[i] and output[0][0] < output[0][1]:
        acc = acc + 1

    numtotal = int(0.2*len(self.dfname.print()))
    acc = (acc/numtotal)*100
    print('Accuracy =',acc,'%')
    modelacc[self.team] = acc
```

```python
def test(testds, param):

  for i in range(len(testds)):

      if testds.toss_choice[i] == 'Bat':
        choice = 1
      else:
        choice = 0

      input = np.array([teams[testds.team1[i]], teams[testds.team2[i]], teams[testds.toss_winner[i]], choice, stadiums[testds.stadium[i]]])

      if modelacc[testds.team1[i]] >= modelacc[testds.team2[i]]:
        w , b = param[testds.team1[i]]
      elif modelacc[testds.team1[i]] < modelacc[testds.team2[i]]:
        w , b = param[testds.team2[i]]

      output = np.matmul(input,w) + b

      if output[0][0] > output[0][1] :
        print(1)
      elif output[0][1] > output[0][0] :
        print(2)
      else:
        print(0) #Tie

```

```python
csk = iterator(CSK, 'Chennai Super Kings')
mi = iterator(MI, 'Mumbai Indians')
kkr = iterator(KKR, 'Kolkata Knight Riders')
pk = iterator(PK, 'Punjab Kings')
dc = iterator(DC, 'Delhi Capitals')
rr = iterator(RR, 'Rajasthan Royals')
srh = iterator(SRH, 'Sunrisers Hyderabad')
kxip = iterator(KXIP, 'Kings XI Punjab')
dd = iterator(DD, 'Delhi Daredevils')
gl = iterator(GL, 'Gujarat Lions')
rps = iterator(RPS, 'Rising Pune Supergiant')
pw = iterator(PW, 'Pune Warriors')
ktk = iterator(KTK, 'Kochi Tuskers Kerala')
rcb = iterator(RCB, 'Royal Challengers Bangalore')
```

```python
csk.train(5)
mi.train(5)
kkr.train(5)
pk.train(5)
dc.train(5)
rr.train(5)
srh.train(5)
kxip.train(5)
dd.train(5)
gl.train(5)
rps.train(5)
pw.train(5)
ktk.train(5)
rcb.train(5)


csk.validate()
mi.validate()
kkr.validate()
pk.validate()
dc.validate()
rr.validate()
srh.validate()
kxip.validate()
dd.validate()
gl.validate()
rps.validate()
pw.validate()
ktk.validate()
rcb.validate()
```
